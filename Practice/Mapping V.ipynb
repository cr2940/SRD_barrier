```python
import numpy as np
import matplotlib.pyplot as plt
m1 = 0.616  # slope of one edge of the V barrier
def m(y):
    # slope function with respect to position y under the V barrier
    y_low = y<=0.72 
    y_abo = y>0.72
    slope = (m1/0.72 * y)*y_low + (-m1/0.28 * y + m1 + (m1/0.28)*0.72)*y_abo
    return slope
def Vmap(xc,yc):
    # the V mapping function
    xp = xc.copy()
    yp = yc.copy()
    yp = yc - np.absolute(np.multiply(m(yc),np.minimum(1-xc,xc)))
    return xp,yp
```

```python
# set computational grid and calculate the mapping
xc = np.linspace(0,1,30)
yc = np.linspace(0,1,30)
XM,YM = np.meshgrid(xc,yc)
xp,yp = Vmap(XM,YM)
xygrid = np.column_stack([[x, y] for x in xp for y in yp])
```

```python
# plot the mapped grid
plt.scatter(xygrid[0], xygrid[1],s=3,edgecolor="none")
plt.grid(True)
plt.axis("equal")
```
